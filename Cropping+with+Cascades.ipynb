```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
cascade = cv2.CascadeClassifier('C:/Users/Peter/Anaconda3/pkgs/opencv-3.3.1-py36h20b85fd_1/Library/etc/haarcascades/haarcascade_frontalface_alt.xml')
img = cv2.imread('43M_SP_O.bmp')
```

```python
def crop(img, method):
    cascade = cv2.CascadeClassifier(method)
    img = cv2.imread(img)
    faces = cascade.detectMultiScale(img)
    for (x,y,w,h) in faces:
        sub_face = img[y:y+h, x:x+w]
        cropped = plt.imshow(cv2.cvtColor(sub_face, cv2.COLOR_BGR2RGB))
    return cropped
```
