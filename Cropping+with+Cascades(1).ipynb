```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
os.chdir('C:/Users/Peter/NimStim/Faces_Uncropped')
```

```python
def crop(img, method):
    imgpath = 'C:/Users/Peter/NimStim/Faces_Uncropped/'+img
    img = cv2.imread(imgpath)
    cascade = cv2.CascadeClassifier(method)
    tempnames = imgpath.split('/')
    name = tempnames[-1]
    img_name = name[0:-4] + '_Cropped'
    faces = cascade.detectMultiScale(img)
    for (x,y,w,h) in faces:
        sub_face = img[y:y+h, x:x+w]
        cropped = plt.imshow(cv2.cvtColor(sub_face, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.savefig('C:/Users/Peter/NimStim/Faces_Cropped/'+img_name+'.jpg',bbox_inches='tight',pad_inches = 0)
```
